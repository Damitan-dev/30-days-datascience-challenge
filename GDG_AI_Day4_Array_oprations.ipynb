#### Numpy Arrays and Operations

## Why use MEAN,STANDARD DEVIATION AND RESHAPING

# Mean
- Gives summary of dataset
-Used in Normalization
-Used in ML algorithms(e.g feature scaling)

# Standard deviation
- Helps detect outliers
- Used in normalization
-Shows staability of data
-Tells which feature varies the most

# Reshaping
- Only changing the arrangement not the data
-It ensure compatibility with linear regression e.t.c.
- When preparing data for training



```python
import numpy as np #Becuase Numpy handles fast numerical  computation using arrays(Vectors,matrices)
import pandas as pd

#Load the data

df = pd.read_csv("cleaned_data.csv")

#Making sure the numeric columns are actually numeric
df[["Duration","Pulse","Maxpulse","Calories"]].info()


```

## Creating 1D Array

The 1D array represents a single numerical feature extracted from the dataset (e.g Calories).

shape: (30,)
It contains 30 values arranged in a single line.
Each value corresponds to one workout session.
This structure isuseful for calculating statistical measures like mean and standard deviation.

```python
# Step 1: Creating 1D array
calories = df["Calories"].to_numpy()

#To check the type if it's an array use-type(calories)
#To check the array dimension use-calories.ndim
#To check the shape of the array use-calories.shape





```

My 1D array represents Calories per session

Calculating the Mean and Standard deviation of the 1D Array

```python

#Calculating the Mean
display(np.mean(calories))

#Calculating the Standard deviation
display(np.std(calories))  #If small consistent workouts,if large inconsistent intensity


#Shaping 1D array
display(calories.reshape(-1,1))
```

#### Creating a 2D Array

The 2D array represents multiple numerical features combined into one matrix.

```python
#Creating 2D array
features = df[["Duration","Pulse","Maxpulse","Calories"]].to_numpy()
display(np.mean(features,0)) #Mean of each column why it is 0

#Standard deviation for 2D array columns
display(np.std(features,0))

#Reshaping 
calories_reshaped = calories.reshape(-1,1)
display(calories_reshaped.shape)

```
