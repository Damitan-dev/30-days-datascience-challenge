#### Visualizing the dataset insights using charts

```python
import pandas as pd
import matplotlib.pyplot as plt

#Load the cleaned dataset
df = pd.read_csv("cleaned_data.csv")


```

### Bar Chart

Bar charts are best for categorical data.
 I can convert Duration into categories if needed

 It compares counts or aggregated valus across categories


```python
df["Duration"].value_counts().plot(kind="bar")

plt.title("Workout Duration frequency")
plt.xlabel("Duration")
plt.ylabel("Count")
plt.show()
```

### Line Chart

It shows change over continuous variables,typically time.It is for trend analysis



```python
#First sort dataset by date so the line moves in correct time order

df = df.sort_values("Date")
plt.plot(df["Date"],df["Calories"]) #Plot Date on x-axis, plot Calories on Y-axis and connect points with a line

plt.title("Calories Burned over Time") # Chart title
plt.xlabel("Date")
plt.ylabel("Calories")
plt.xticks(rotation=85) #Rotate date labels for better readability
plt.show()
```

### Histogram

It shows the distributiion of numeric values

```python
df["Pulse"].plot(kind="hist", bins=15) #Select the pulse column,plot an histogram of it,then bins=15 divides pulse values into 15 intervals


plt.title("Distribution of pulse")
plt.xlabel("Pulse")
plt.ylabel("Frequency") #(How many times vallues occur in each bin)

plt.show() #Display the plot
```
