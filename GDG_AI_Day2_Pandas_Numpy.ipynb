```python
import pandas as pd
import numpy as np

#Step 1:Load Dataset

df = pd.read_csv("data 2.csv")
df.head()


```

```python
#Step 2: Check duplicates

#Detect duplicates
duplicates = df.duplicated().sum()
print("Duplicates:",duplicates)

df = df.drop_duplicates()

print("Duplicates after removal:",df.duplicated().sum())
print("Shape:",df.shape)
```

```python
#step 3:Drop or fill missing values

df.isnull().sum()

#Fill Calories  with Median, why median? Because Calorie data can be skw=ewed.Median is more roc=bust to outliers.


median_calories = df["Calories"].median()
df["Calories"] = df["Calories"].fillna(median_calories)


#Drop row with Missing  Date
#A missing data  is ambigious and harder to reconstruct logically

df = df.dropna(subset=["Date"])

# Verifying if what I did above worked
display(df.isnull().sum())
df.shape
```
